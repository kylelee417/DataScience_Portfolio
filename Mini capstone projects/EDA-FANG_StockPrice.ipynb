# Background

This capstone project explores an exploratory data analysis of stock prices. The dataset is fetched from online (*Yahoo stocks*), and the past 5 year's of stock trends of FANG (Facebook, Amazon, Netflix and Google) are explored.

# Data fetching and overview

```python
import pandas_datareader.data as web
import pandas as pd
import numpy as np
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
start = datetime.datetime(2013,1,1)
end = datetime.datetime(2018,1,1)
FANG = ['FB','AMZN','NFLX','GOOG']
```

```python
for stock in FANG:
    vars()[stock] = web.DataReader(stock,'yahoo',start,end)
    print('{}: {}'.format(stock,vars()[stock].shape))
```

```python
df_FANG = pd.concat([FB,AMZN,NFLX,GOOG],keys=FANG,axis=1)
df_FANG.columns.names = ['FANG','Stock Info']
```

```python
df_FANG.head()
```

# EDA
#### Maximul close price for each company's stock for the past 5 years

```python
df_FANG.xs(key='Close',level='Stock Info',axis=1).max()
```

#### Return calculation
$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns = pd.DataFrame()
```

```python
for comp in FANG:
    returns[comp+' Return'] = df_FANG[comp]['Close'].pct_change()
```

```python
returns
```

#### Pairplot for return

```python
pp = sns.pairplot(returns[1:])
```

#### The date of best and worst return for each company

```python
df_max = pd.concat([returns.idxmax(),(returns.max()*100).round(2)],axis=1)
df_max.columns=['Date','Return %']
df_max
```

```python
df_min = pd.concat([returns.idxmin(),(returns.min()*100).round(2)],axis=1)
df_min.columns=['Date','Return %']
df_min
```
