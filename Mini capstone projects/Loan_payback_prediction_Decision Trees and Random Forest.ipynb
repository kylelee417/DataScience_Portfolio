# Background

This project will explore data from [LendingClub.com](www.lendingclub.com). Lending Club connects people who need money (borrowers) with people who have money (investors). 

The goal of this project is to predict if borrower will pay back the loan in full based on their profiles. This dataset contains lending data from 2007-2010. 

Data columns include:
* credit.policy: 1 if the customer meets the credit underwriting criteria of LendingClub.com, and 0 otherwise.
* purpose: The purpose of the loan (takes values "credit_card", "debt_consolidation", "educational", "major_purchase", "small_business", and "all_other").
* int.rate: The interest rate of the loan, as a proportion (a rate of 11% would be stored as 0.11). Borrowers judged by LendingClub.com to be more risky are assigned higher interest rates.
* installment: The monthly installments owed by the borrower if the loan is funded.
* log.annual.inc: The natural log of the self-reported annual income of the borrower.
* dti: The debt-to-income ratio of the borrower (amount of debt divided by annual income).
* fico: The FICO credit score of the borrower.
* days.with.cr.line: The number of days the borrower has had a credit line.
* revol.bal: The borrower's revolving balance (amount unpaid at the end of the credit card billing cycle).
* revol.util: The borrower's revolving line utilization rate (the amount of the credit line used relative to total credit available).
* inq.last.6mths: The borrower's number of inquiries by creditors in the last 6 months.
* delinq.2yrs: The number of times the borrower had been 30+ days past due on a payment in the past 2 years.
* pub.rec: The borrower's number of derogatory public records (bankruptcy filings, tax liens, or judgments).

# Feature engineering
Since `purpose` is categorical, it should be numerically encoded. There are 7 categoreis in `purpose`, therefore, it can be easily one-hot-encoded.

```python
df_loans['purpose'].nunique()
```

```python
df_loans_final = pd.get_dummies(df_loans,columns=['purpose']) 
```

```python
df_loans_final.head()
```

# Model training
#### train and test split

```python
X = df_loans_final.drop('not.fully.paid',axis=1)
y = df_loans_final['not.fully.paid']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=100)
```

## model #1: Decision Tree

```python
dtc = DecisionTreeClassifier().fit(X_train,y_train)
```

## Decision Tree prediction and evaluation

```python
y_predict = dtc.predict(X_test)
```

```python
pd.DataFrame(confusion_matrix(y_test,y_predict),index=['True fully paid','True not fully paid'],
             columns=['Predicted fully paid','Predicted not fully paid'])
```

```python
print(classification_report(y_test,y_predict))
```

Without any hypermatermeter tuning, the base decision tree model showed ~0.74 f1-score.

## model #2: Random Forest 

```python
rfc = RandomForestClassifier(n_estimators=100).fit(X_train,y_train)
```

## Random Forest prediction and evaluation

```python
y_predict = rfc.predict(X_test)
```

```python
pd.DataFrame(confusion_matrix(y_test,y_predict),index=['True fully paid','True not fully paid'],
             columns=['Predicted fully paid','Predicted not fully paid'])
```

```python
print(classification_report(y_test,y_predict))
```

Without any hypermatermeter tuning, the base random forest model showed ~0.76 f1-score, a small increase in performance compared to the single decision tree model. In more details, the random forest model could dramatically reduce the number of False Positive. While not being explored, by tuning the hyperparameter, the performance of the random forest is expected to improve further.
